```python
import matplotlib.pyplot as plt

countries = ['Iran', 'India', 'Spain', 'Afghanistan', 'Others']
production = [90, 5 , 2 , 2 , 1]

plt.figure(figsize=(8, 6))
plt.pie(production, labels=countries, autopct='%1.0f%%', startangle=100,pctdistance = 0.8 )
plt.title('Global Saffron Production Share')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
explode = [0.05] * len(production)
plt.pie(production, labels=countries, autopct='%1.0f%%', pctdistance=0.85, explode=explode, shadow=True)
plt.title('Saffron Export Market Share')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#   2024   2025
months = pd.date_range(start='2024-06-01', periods=12, freq='M')

#     
global_low = np.full(12, 1166.5)
global_high = np.full(12, 1952.4)

#    ( 11001510 USD/kg)
iran = [1100, 1150, 1200, 1250, 1300, 1350, 1400, 1450, 1500, 1510, 1450, 1400]

plt.figure(figsize=(12, 6))
plt.fill_between(months, global_low, global_high, color='blue', alpha=0.2, label='Global Export Range')
plt.plot(months, iran, color='green', linewidth=2, label='Iran Bulk Export Price')

plt.title(' Saffron Export Price: Global vs Iran (Jun2024May2025)')
plt.xlabel('Month')
plt.ylabel('Price (USD/kg)')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
