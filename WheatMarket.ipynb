```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Download wheat data from Yahoo Finance (ZW=F = Wheat Futures)
wheat = yf.download('ZW=F', start='2019-01-01', end='2024-12-31')

# Clean and keep only 'Close' column
wheat = wheat.dropna()
wheat = wheat[['Close']]

# Define the start of the Ukraine war
war_start = '2022-02-24'

# Split data into before and after the war
before_war = wheat[wheat.index < war_start]
after_war = wheat[wheat.index >= war_start]

# Calculate percentage price changes
change_before = float((before_war['Close'].iloc[-1] - before_war['Close'].iloc[0]) / before_war['Close'].iloc[0] * 100)
change_after = float((after_war['Close'].iloc[-1] - after_war['Close'].iloc[0]) / after_war['Close'].iloc[0] * 100)

# Print results
print(f"Price change before the war: {change_before:.2f}%")
print(f"Price change after the war started: {change_after:.2f}%")

# Plot the wheat price
plt.figure(figsize=(12, 6))
plt.plot(wheat.index, wheat['Close'], label='Wheat Price')
plt.axvline(pd.to_datetime(war_start), color='red', linestyle='--', label='War Start (Feb 24, 2022)')
plt.title('Wheat Price Before and After Ukraine War')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
